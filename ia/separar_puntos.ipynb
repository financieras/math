```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Fijar la semilla para reproducibilidad
random.seed()

# Crear la matriz de puntos aleatorios
matrix = []
for i in range(1000):
    x = random.random() * 10 - 5
    y = random.random() * 10 - 5
    matrix.append([x, y])

# Convertir la matriz a un arreglo NumPy
matrix = np.array(matrix)

# Separar las coordenadas x e y
x_coords = matrix[:, 0]
y_coords = matrix[:, 1]

# Crear la gráfica de dispersión
plt.scatter(x_coords, y_coords, alpha=0.3)
plt.title('Nube de puntos aleatorios')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
